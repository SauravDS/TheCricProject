```python
##Terminologies

#Matches & Innings

#Matches : included in playing XI
#Innings : If the batman comes to bat in a mach he is playing

#  bat_avg = Batting Average : Runs/dismissals
# bat_str = Batting Stikerate : (runs/balls)*100
# rpi = Runs per innings : (Runs/innings)
# bpb = Balls per boundary : (balls / boundaries)
#dot_percentage : Percentage of dot balls played by  a batsman in his entire innings : (dot_balls/balls)*100
#non_boundary_percentage :
#boundary_percentage = (boundary_balls/balls)*100
#non-boundary percentage = (balls-balls_boundary)/(balls)*100
```

```python
#Example for understding terminologies

batsman = 'Suryakumar Yadav'
runs_scored = [45,32]
balls_faced = [60,46]
dismissals = 1
innings = 2
runs = 77
balls = 106
dot_balls = 28
boundary_balls = 7
fours = 4
sixes = 3
```

```python
bat_avg = runs/dismissals
rpi = runs/innings
bat_str = (runs/balls)*100
boundaries = fours + sixes
bpb = balls/boundaries
dot_percentage = (dot_balls/balls)*100
boundary_percentage = (boundary_balls/balls)*100
non_boundary_percentage = (balls-boundary_balls)/(balls)*100
```
